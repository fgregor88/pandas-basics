### Pandas build in selector operators (row-first column-second)

```python
reviews.iloc[0] #first row
```

```python
reviews.iloc[:,1] #second column (the first one is just the index)
```

```python
reviews.iloc[:,1:4] # all rows and columns from 1 to 4
```

```python
reviews.iloc[:,[1,3,5,6]] # all rows and columns 1 3 5 and 6
```

```python
reviews.iloc[-5:] # last 5 rows
```

### Label selector

```python
reviews.loc[0,'country'] # first row, column 'country'
```

```python
reviews.loc[0:10, ['country', 'points', 'province']] # first 11 rows, columns 'country', 'points' and 'province'
```

```python
reviews.set_index('title')
```

```python
reviews.country == 'Italy'
```

```python
reviews.loc[reviews.country == 'Italy']
```

```python
reviews.loc[(reviews.country == 'Italy') & (reviews.points > 90) ]
```

```python
reviews.loc[(reviews.country == 'France') | (reviews.points >= 90)]
```

```python
reviews.loc[reviews.country.isin(['France','Italy'])]
```

```python
reviews.loc[reviews.price.notnull()]
```

```python
reviews['critic'] = 'everyone'
reviews['critic']
```

```python
reviews['index_backwards'] = range(len(reviews), 0, -1)
reviews['index_backwards']
```
